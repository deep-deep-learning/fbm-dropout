```python
import matplotlib.pyplot as plt
from stochastic.processes.continuous import FractionalBrownianMotion


fbm_x = FractionalBrownianMotion(hurst=0.9, t=1)
fbm_y = FractionalBrownianMotion(hurst=0.9, t=1)
x_1 = fbm_x.sample(100)
y_1 = fbm_y.sample(100)

x_1 %= 1
y_1 %= 1

for i in range(10):
    plt.scatter(x_1[i*10:i*10+10], y_1[i*10:i*10+10])
    plt.xlim(0, 1)
    plt.ylim(0, 1)
    plt.show()
```
